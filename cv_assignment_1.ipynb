###1.Importing Necessary Libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import skimage as io
import cv2
from google.colab.patches import cv2_imshow
```

###2.Taking input of both Image and converting it into Grayscale.

```python
image_1 = cv2.imread("/content/Image_1.jpeg")
image_1 = np.array(image_1)
image_1_gray = cv2.cvtColor(image_1,cv2.COLOR_BGR2GRAY)

image_2 = cv2.imread("/content/Image_2.jpeg")
image_2 = np.array(image_2)
image_2_gray = cv2.cvtColor(image_2,cv2.COLOR_BGR2GRAY)
```

###4.Local Binary Pattern Function.

```python
def calculate_lbp(img, x, y):
    center = img[x][y]
    values = []
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue
            values.append(1 if img[x + i][y + j] >= center else 0)
    val = 0
    for i in range(len(values)):
      val+=values[i]*2**i

    return val
```

```python
def apply_lbp(img):
    height, width = img.shape
    lbp_img = np.zeros((height, width), dtype=np.uint8)
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            lbp_img[i, j] = calculate_lbp(img, i, j)
    return lbp_img
```

###5. Calling of Function.

```python
lbp_image_1 = apply_lbp(image_1_gray)
lbp_image_2 = apply_lbp(image_2_gray)
```

###6. Displaying Result.

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
ax[0].imshow(lbp_image_1, cmap='gray')
ax[0].set_title('lbp_image_1')
ax[1].imshow(lbp_image_2, cmap='gray')
ax[1].set_title('lbp_image_2')
plt.show()
```
